```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('OscarsDatasets.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Plot histograms for each variable
sns.histplot(df['year_film'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribution of year_film')

sns.histplot(df['year_ceremony'], bins=30, kde=True, ax=axes[1])
axes[1].set_title('Distribution of year_ceremony')

sns.histplot(df['ceremony'], bins=30, kde=True, ax=axes[2])
axes[2].set_title('Distribution of ceremony')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[['year_film', 'year_ceremony', 'ceremony']], diag_kind='kde')
plt.show()

# Bar chart for the count of winners
sns.countplot(x='winner', data=df)
plt.show()

# Unique values in the 'category' column
print(df['category'].value_counts())

# Check for missing values
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='category', data=df, order=df['category'].value_counts().index)
ax.set_xticklabels(ax.get_xticklabels(), rotation='vertical', ha="right", fontsize=8)  # Adjust rotation angle and font size
plt.title('Distribution of Categories')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='category', hue='winner', data=df, order=df['category'].value_counts().index)
plt.xticks(rotation=90, fontsize=8)
plt.title('Distribution of Categories by Winner')
plt.show()
```
