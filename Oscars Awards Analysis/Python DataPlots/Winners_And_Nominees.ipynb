
## Load the data you receive into a Pandas DataFrame

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Count Plot of Nominee Categories:


```python
plt.figure(figsize=(20, 6))
sns.countplot(data=df[:500], x='category')

plt.xlabel('Nominee Category')
plt.ylabel('Count')
plt.title('Count of Nominees by Category')

plt.xticks(rotation=90)

plt.show()

```

## Bar Plot of Winners by Year:

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df[:500], x='year_film', hue='winner')

plt.xlabel('Year of Film')
plt.ylabel('Count')
plt.title('Winners and Nominees by Year')

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='year_film')

plt.xlabel('Year of Film')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Nominations by Year')

plt.show()
```

## Violin Plot of Nominations by Year:

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, y='year_film')

plt.ylabel('Year of Film')
plt.title('Distribution of Nominations by Year')

plt.show()
```
