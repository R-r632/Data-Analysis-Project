# Tesla and GameStop Stock vs Revenue Dashboard (YFinance Revenue)

## Question 1: Extract Tesla Stock Data Using yfinance

```python
import yfinance as yf
import pandas as pd

tesla_data = yf.download('TSLA', start='2010-06-29', end='2025-04-25')
tesla_data.reset_index(inplace=True)

tesla_data.head()
```

## Question 2: Extract Tesla Revenue Data Using yfinance Financials

```python

tsla_ticker = yf.Ticker('TSLA')
tsla_fin = tsla_ticker.financials

tesla_revenue = tsla_fin.loc['Total Revenue'].reset_index()
tesla_revenue.columns = ['Date', 'Revenue']

tesla_revenue.tail()
```

## Question 3: Extract GameStop Stock Data Using yfinance

```python
gme_data = yf.download('GME', start='2002-08-13', end='2025-04-25')
gme_data.reset_index(inplace=True)

gme_data.head()
```

## Question 4: Extract GameStop Revenue Data Using yfinance Financials

```python
gme_ticker = yf.Ticker('GME')
gme_fin = gme_ticker.financials

gme_revenue = gme_fin.loc['Total Revenue'].reset_index()
gme_revenue.columns = ['Date', 'Revenue']

gme_revenue.tail()
```

## Questions 5 & 6: Plot Dashboards

```python
import plotly.graph_objects as go

def make_graph(stock_df, revenue_df, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_df['Date'], y=stock_df['Close'], name='Close Price'))
    fig.add_trace(go.Bar(x=revenue_df['Date'], y=revenue_df['Revenue'], name='Revenue', yaxis='y2'))
    fig.update_layout(
        title=title,
        xaxis_title='Date',
        yaxis=dict(title='Stock Price (USD)'),
        yaxis2=dict(title='Revenue (USD)', overlaying='y', side='right')
    )
    fig.show()

# Plot Tesla
make_graph(tesla_data, tesla_revenue, 'Tesla Stock Price vs Revenue')

# Plot GameStop
make_graph(gme_data, gme_revenue, 'GameStop Stock Price vs Revenue')
```
